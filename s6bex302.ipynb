## 爬取網頁中的所有圖片, 放入output資料夾

```python
import requests
import urllib
import os
from PIL import Image
from bs4 import BeautifulSoup  

url = 'http://www.macaodaily.com/html/2021-05/12/content_1515629.htm'

r = requests.get(url)
r.encoding="utf-8"
html=r.text
print(html)
### 取得網頁的html

bs = BeautifulSoup(html, 'lxml') 
elements = bs.find_all("img")
print(elements)
### 取得所有elements

for item in elements:
    src = item.get('src')
    link = urllib.parse.urljoin(url,src)
    print(link)
    im = Image.open(urllib.request.urlopen(link))
    
    filename = os.path.basename(link)
    im.save(os.path.join('output',filename))
    


    
    
```
