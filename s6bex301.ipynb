```python
import requests
from bs4 import BeautifulSoup  

r = requests.get('http://std.puiching.edu.mo/~ricky/pcms.html')
r.encoding="utf-8"
html=r.text
print(html)

bs = BeautifulSoup(html, 'lxml') 
elements = bs.find_all("p")
print(elements)
for item in elements:
    print(item.get_text())
    
```

```python
import requests
from bs4 import BeautifulSoup  
import os
import urllib.parse


url = 'http://www.puiching.edu.mo/main/'
base= url
r = requests.get(url)
r.encoding="utf-8"
html=r.text
print(html)

bs = BeautifulSoup(html, 'lxml') 
elements = bs.find_all("a")
print(elements)
for item in elements:
    link = item.get('href')
    link = urllib.parse.urljoin(url,link)
    print(link)
    
    
```
